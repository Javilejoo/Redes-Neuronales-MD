#### Distribución de precios de venta

```python
# Extraer los precios de venta
precios_venta = df_train['SalePrice']

# Crear el histograma
plt.figure(figsize=(10, 6))
plt.hist(precios_venta, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribución de Precios de Venta de Casas')
plt.xlabel('Precio de Venta')
plt.ylabel('Frecuencia')
plt.grid(True)
```

#### Correlación entre características y precios de venta
- una matriz de correlación que muestra la relación entre las diferentes característica

```python
# Seleccionar las características numéricas para calcular la correlación
features = ['MSSubClass', 'LotFrontage', 'LotArea', 'OverallQual', 'OverallCond',
            'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2',
            'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF',
            'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',
            'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces',
            'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF',
            'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal',
            'MoSold', 'YrSold']

# Calcular la matriz de correlación
correlation_matrix = df_train[features + ['SalePrice']].corr()

# Visualizar la matriz de correlación utilizando un mapa de calor
plt.figure(figsize=(20, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlación entre Características y Precio de Venta')
```

En base  a los precios agregaremos una nueva columna que determina el tipo de precio categoria de la casa 

```python
def categorize_price(price):
    if price < 150000:
        return 'Cheap'
    elif price < 300000:
        return 'Medium'
    else:
        return 'Expensive'

# Aplicar la función para crear la nueva columna
df_train['PriceCategory'] = df_train['SalePrice'].apply(categorize_price)
```

### Encoding
Encodiaremos las categorias con numeros para entrenar el modelo. 
* 'Cheap': 0
* 'Medium': 2
* 'Expensive': 1

```python
from sklearn.preprocessing import LabelEncoder

# Inicializar el codificador
label_encoder = LabelEncoder()

# Aplicar el Label Encoding a la columna 'PriceCategory' en df_train
df_train['PriceCategoryEncoded'] = label_encoder.fit_transform(df_train['PriceCategory'])

# Ver los valores únicos y su correspondiente encoding
print(df_train[['PriceCategory', 'PriceCategoryEncoded']].drop_duplicates())
```
